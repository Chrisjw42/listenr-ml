# Fourier Transforms

Mathematically, a Fourier Tranform is an integral of the form:

$$\int_{t_1}^{t_2} g(t) \cdot e^{-2\pi i f t} \; dt$$

where $f$ is the frequency of $g(t)$ ,

$t$ is some instance of time ,

$e^{-2\pi i  t}$ is the complex exponential, traversed in the clockwise direction

and $g(t)$ is our original function we wish to decompose.

For example:

Let's say we have a function $g(t) = cos(2x+1) + 3sin(4x) + sin(8x)$

Looking at that, it's evident that the component waveforms are $g_1(t) = cos(2x+1)$, $g_2(t) = 3sin(4x)$ and $g_3(t) = sin(8x)$.


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0,4*np.pi-1,0.1)
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(17, 15))
q = np.cos(2*x+1)
r = np.sin(4*x)*3
s = np.sin(8*x)


ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].plot(x, q, label='$g_1(t) = cos(2x+1)$')
ax[0].plot(x, r, label='$g_2(t) = 3sin(4x)$')
ax[0].plot(x, s, label='$g_3(t) = sin(8x)$')
ax[0].legend(fontsize=16, loc=4)
ax[0].set_ylim(-5,4,1)

ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].plot(x, q+r+s, label='$g(t) = cos(2x+1) + 3sin(4x) + sin(8x)$')
ax[1].legend(fontsize=16, loc=4)
ax[1].set_ylim(-5,4,1)

```

$g(t)$ has a frequency of $2$.  

Taking the Fourier transform of $g(t)$ for the period of $[0,2\pi]$:

$$\int_0^{2\pi} \bigg[cos(2x +1) + 3sin(4x) + sin(8x) \bigg] \cdot e^{-4\pi i t} \; dt$$
$$= \frac{1}{2\pi} \bigg[e^{-4i\pi^2}sin(4\pi^2)\big(cos(2x +1) + 3sin(4x) + sin(8x)\big) \bigg]$$

Graphing this:

```python
t = q+r+s
u = -4*1.j*np.pi*np.pi
o = np.exp(u)
v = np.sin(4*np.pi*np.pi)
w = 1./(2*np.pi)

z = o*v*t
a = w*z


x = np.arange(0,4*np.pi-1,0.1)

fig, ax = plt.subplots(figsize=(17, 15))
ax.set_ylim(-5,4,1)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, a)
ax.plot(x, q, label='$g_1(t) = cos(2x+1)$')
ax.plot(x, r, label='$g_2(t) = 3sin(4x)$')
ax.plot(x, s, label='$g_3(t) = sin(8x)$')
ax.legend(fontsize=16, loc=4)
```
